```python
"""
    get_exturlusage.py

    MediaWiki API Demos
    Demo of `Exturlusage` module: Enumerate pages that contain a given URL.

    MIT License
"""

import requests

S = requests.Session()

def get_links_from_wikis(code,repo_url):
    URL = f"https://{code}.wikipedia.org/w/api.php"

    PARAMS = {
        "action": "query",
        "format": "json",
        "list": "exturlusage",
        "euquery": repo_url,
        "eulimit" : 500,
        "meta" : "siteinfo"
    }

    R = S.get(url=URL, params=PARAMS)
    print(R.url)
    DATA = R.json()

    EXTURLS = DATA["query"]["exturlusage"]

    return_list = []
    for exturl in EXTURLS:
        print("Page " + exturl["title"] + " has " + exturl["url"] + " url.")
        return_list.append({'repo' : repo_url, 'titel': exturl["title"], 'url' : exturl["url"], 'code' : code, 'pageID' : exturl["pageid"]})
    return return_list
```

Erzeuge eine Liste der verschiedenen Wikipedias auf Basis einer entsprechenden Liste

```python
wikipedia_list_url = "https://en.wikipedia.org/wiki/List_of_Wikipedias"

req = requests.get(wikipedia_list_url)
html = req.content
soup = BeautifulSoup(html, 'html.parser')
table = soup.findAll('table')[1]
```

```python
WIKIS = []
rows = table.findAll('tr')
for row in rows[1:]:
    link = row.findAll('td')[2].find('a')
    code = link.text
    wiki_link = link['href']
    language = row.find('td').text
    print(code,wiki_link, language)
    WIKIS.append({
        'code' : code,
        'link' : wiki_link,
        'lang' : language
    })
```

```python
OUTPUT_DICT = {}
for name,url in URLS.items():
    output = []
    for wiki in WIKIS:
        print(f"{wiki['lang']}\n====")
        try:
            OUT = get_links_from_wikis(wiki['code'],url)
            print(OUT)
            output.append(OUT)
        except Exception as e:
            print(e)
            print(wiki['lang'])
    #flat_list = [item for sublist in output for item in sublist]
    #language_distribution = {}
    #for elem in flat_list:
    #    try:
    #        language_distribution[elem['code']].append(elem['url'])
    #    except Exception as e:
    #        language_distribution[elem['code']] = [elem['url']]
    #DICT = sorted(language_distribution.items(), key=lambda x: len(x[1]), reverse=True)
    #xs = [x[0] for x in DICT]
    #ys = [len(x[1]) for x in DICT]
    #plt.bar(xs,ys,label=name)
    #plt.xticks(rotation=90)
    #plt.legend()
    #plt.tight_layout()
    #titel = f"{name}-Links auf Wikipedia"
    #plt.title(titel)
    #plt.savefig(f"{name}_Links_Wiki.png",savefig=250,bbox_inches='tight')
    #plt.clf()
    OUTPUT_DICT[name] = output
```

```python
PLOT_DICT = {}

for k in OUTPUT_DICT.keys():
    if not k in PLOT_DICT.keys():
        PLOT_DICT[k] = {}
    for elem in OUTPUT_DICT[k]:
        for e in elem:
            try:
                PLOT_DICT[k][e['code']] += 1
            except KeyError:
                PLOT_DICT[k][e['code']] = 1
```
